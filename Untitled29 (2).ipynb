#Analysis for catagorical varibale

```python
class CategoricalPlotter:
    def __init__(self, dataframe, palette='viridis'):
        self.df = dataframe
        self.palette = palette

    def bar_plot(self, variable):
        plt.figure(figsize=(9, 3))
        sns.countplot(x=variable, data=self.df)
        plt.title(f'Bar Plot of {variable}')
        plt.xlabel(variable)
        plt.ylabel('Frequency')
        plt.show()
        print(f"{variable}: \n{self.df[variable].value_counts()}")

    def pie_plot(self, variable):
        plt.figure(figsize=(8, 8))
        counts = self.df[variable].value_counts()
        plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette(self.palette, len(counts)))
        plt.title(f'Pie Plot of {variable}')
        plt.axis('equal')
        plt.show()

plotter = CategoricalPlotter(df, palette='viridis') # Using viridis as an example of a visually appealing palette

categorical_vars = [ 'Parch', 'SibSp']



```

```python
print("\nGenerating Pie Plots:")
for var in categorical_vars:
    plotter.pie_plot(var)
```

```python
# @title servival-Class-Child/Adult/Senior

def age_group(age):
    if age <= 12:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['Age'].apply(age_group)



survived_df = df[df['Survived'] == 1].copy()
not_survived_df = df[df['Survived'] == 0].copy()
survived_child_df = survived_df[survived_df['AgeGroup'] == 'Child'].copy()
survived_adult_df = survived_df[survived_df['AgeGroup'] == 'Adult'].copy()
survived_senior_df = survived_df[survived_df['AgeGroup'] == 'Senior'].copy()

not_survived_child_df = not_survived_df[not_survived_df['AgeGroup'] == 'Child'].copy()
not_survived_adult_df = not_survived_df[not_survived_df['AgeGroup'] == 'Adult'].copy()
not_survived_senior_df = not_survived_df[not_survived_df['AgeGroup'] == 'Senior'].copy()
survived_child_pclass_counts = survived_child_df['Pclass'].value_counts()
survived_adult_pclass_counts = survived_adult_df['Pclass'].value_counts()
survived_senior_pclass_counts = survived_senior_df['Pclass'].value_counts()

not_survived_child_pclass_counts = not_survived_child_df['Pclass'].value_counts()
not_survived_adult_pclass_counts = not_survived_adult_df['Pclass'].value_counts()
not_survived_senior_pclass_counts = not_survived_senior_df['Pclass'].value_counts()


import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=3,
                    subplot_titles=('Survived - Child', 'Survived - Adult', 'Survived - Senior',
                                      'Not Survived - Child', 'Not Survived - Adult', 'Not Survived - Senior'),
                    specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}],
                           [{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]])

fig.add_trace(go.Pie(labels=survived_child_pclass_counts.index, values=survived_child_pclass_counts.values), 1, 1)
fig.add_trace(go.Pie(labels=survived_adult_pclass_counts.index, values=survived_adult_pclass_counts.values), 1, 2)
fig.add_trace(go.Pie(labels=survived_senior_pclass_counts.index, values=survived_senior_pclass_counts.values), 1, 3)
fig.add_trace(go.Pie(labels=not_survived_child_pclass_counts.index, values=not_survived_child_pclass_counts.values), 2, 1)
fig.add_trace(go.Pie(labels=not_survived_adult_pclass_counts.index, values=not_survived_adult_pclass_counts.values), 2, 2)
fig.add_trace(go.Pie(labels=not_survived_senior_pclass_counts.index, values=not_survived_senior_pclass_counts.values), 2, 3)

fig.update_layout(title_text='Pclass Distribution by Survival Status and Age Group')

fig.show()
```

# Data Cleaning & Preprocessing

one of the importent col the age col has notible missing value just replacing the missing value with mean meddin or mode is not best option because servival may be high correlation with age

```python
# @title Handal null values
# prompt: create heatmap to indentified missing value in df
df['Age'] = df.groupby('Survived')['Age'].transform(lambda x: x.fillna(x.median()))
df = df.drop('Cabin', axis=1)

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Calculate the average age for each survival group
average_age_by_survived = df.groupby('Survived')['Age'].mean().reset_index()

# Plot the average age by survival group as a line plot
plt.figure(figsize=(8, 5))
sns.lineplot(x='Survived', y='Age', data=average_age_by_survived, marker='o')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.title('Average Age by Survival Group')
plt.xlabel('Survival Status')
plt.ylabel('Average Age')
plt.grid(True)
plt.show()
```
